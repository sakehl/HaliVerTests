```python
y = 1
yi = 2
mini = (2 if 0 < yi else 0) + y * 8 + yi # mini = (0 < yi) ? 2 : 0 + y * 8 + yi
maxi = mini + (1 if 0 < yi else 3)
print(y, yi, mini, maxi)
```

```python
xy = mini + 1 # mini <= xy <= maxi
print(xy)
```

```python
xxf_xyf_min = 1 + 1026 * mini
xxf_xyf_max = 1 + 1026 * mini + 1026 * (xy - mini)
print(xxf_xyf_min, xxf_xyf_max, xxf_xyf_max-xxf_xyf_min)
```

```python


for xxf_xyf in range(xxf_xyf_min, xxf_xyf_max):
    res = ((xxf_xyf - (1 + 1026 * mini)) // 1026 + mini) % 4 *1024 + (xxf_xyf - (1 + 1026 * mini)) % 1026
    i1 = xxf_xyf
    i2 = ((xxf_xyf - (1 + 1026 * mini)) // 1026 + mini) * 1026 + (xxf_xyf - (1 + 1026 * mini)) % 1026 + 2
    i3 = ((xxf_xyf - (1 + 1026 * mini)) // 1026 + mini) * 1026 + (xxf_xyf - (1 + 1026 * mini)) % 1026
    if((xxf_xyf - (1 + 1026 * mini)) % 1026 < 1024):
        print(f"_blur_x[{res}] = input[{i1}] + input[{i2}] + input[{i3}]")
    else:
        print(f"Not {res}")
```
